```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/raw/Students Social Media Addiction.csv")

plt.figure()
sns.scatterplot(
    x="Avg_Daily_Usage_Hours",
    y="Sleep_Hours_Per_Night",
    data=df
)
plt.title("Daily Social Media Usage vs Sleep Duration")
plt.show()

```

```python
# Create usage categories
df["Usage_Category"] = pd.cut(
    df["Avg_Daily_Usage_Hours"],
    bins=[0, 2, 5, df["Avg_Daily_Usage_Hours"].max()],
    labels=["Low", "Moderate", "High"]
)

# Compute average addiction score per category
usage_summary = df.groupby("Usage_Category")["Addicted_Score"].mean()

usage_summary.plot(kind="bar")
plt.title("Average Addiction Score by Usage Category")
plt.ylabel("Average Addiction Score")
plt.show()

usage_summary
```
